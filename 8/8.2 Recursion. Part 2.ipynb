–¢–µ–º–∞ —É—Ä–æ–∫–∞: —Ä–µ–∫—É—Ä—Å–∏—è
–†–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á –Ω–∞ —Ä–µ–∫—É—Ä—Å–∏—é
–ê–Ω–Ω–æ—Ç–∞—Ü–∏—è. –£—Ä–æ–∫ –ø–æ—Å–≤—è—â–µ–Ω —Ä–µ–∫—É—Ä—Å–∏–∏.

–†–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á –Ω–∞ —Ä–µ–∫—É—Ä—Å–∏—é

–ó–∞–¥–∞—á–∞ 1. –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é draw_rect() —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Ä–µ–∫—É—Ä—Å–∏–∏, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –¥–≤–∞ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ –≤ —Å–ª–µ–¥—É—é—â–µ–º –ø–æ—Ä—è–¥–∫–µ:

width ‚Äì —à–∏—Ä–∏–Ω–∞ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–∞
height ‚Äì –≤—ã—Å–æ—Ç–∞ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–∞

–∏ –ø–µ—á–∞—Ç–∞–µ—Ç –∑–≤–µ–∑–¥–Ω—ã–π –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ —Ä–∞–∑–º–µ—Ä–∞–º–∏ width √ó height.

–†–µ—à–µ–Ω–∏–µ. –î–ª—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Ä–µ–∫—É—Ä—Å–∏–∏ –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ç—Ä–µ—Ç–∏–π –∞—Ä–≥—É–º–µ–Ω—Ç step.

```python
def draw_rect(width, height, step):
    if step < height:
        print('*' * width)
        draw_rect(width, height, step + 1)
```

```python
draw_rect(4, 3, 0)
print()
draw_rect(6, 6, 0)
print()
draw_rect(10, 2, 0)
```

–ù–µ—Å–ª–æ–∂–Ω–æ –∑–∞–º–µ—Ç–∏—Ç—å, —á—Ç–æ —Ç–∞–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ö–æ—Ç—å –∏ –≤—ã–ø–æ–ª–Ω—è–µ—Ç –ø–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—É—é –∑–∞–¥–∞—á—É, –Ω–æ —Å–æ–¥–µ—Ä–∂–∏—Ç –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä step, –∫–æ—Ç–æ—Ä—ã–π –Ω–µ–æ–±—Ö–æ–¥–∏–º –¥–ª—è –æ—Å—Ç–∞–Ω–æ–≤–∫–∏ —Ä–µ–∫—É—Ä—Å–∏–∏. –ú—ã –≤—ã–∑—ã–≤–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é, –ø–æ–¥—Å—Ç–∞–≤–ª—è—è –≤–º–µ—Å—Ç–æ step –∑–Ω–∞—á–µ–Ω–∏–µ 0.

–î–ª—è —Ç–æ–≥–æ —á—Ç–æ–±—ã –∏–∑–±–∞–≤–∏—Ç—å—Å—è –æ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ step –∏ –æ—Å—Ç–∞–≤–∏—Ç—å —Ç–æ–ª—å–∫–æ width –∏ height, –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–≤–∞ –ø–æ–¥—Ö–æ–¥–∞.

–ü–æ–¥—Ö–æ–¥ 1. –°–¥–µ–ª–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é –¥–ª—è step, —Ä–∞–≤–Ω—ã–º –Ω—É–ª—é:

```python
def draw_rect(width, height, step=0):
    if step < height:
        print('*' * width)
        draw_rect(width, height, step + 1)
```

–ú–∏–Ω—É—Å–æ–º —Ç–∞–∫–æ–≥–æ –ø–æ–¥—Ö–æ–¥–∞ —è–≤–ª—è–µ—Ç—Å—è —Ç–æ—Ç —Ñ–∞–∫—Ç, —á—Ç–æ –≤—ã–∑—ã–≤–∞—é—â–∏–π —Ñ—É–Ω–∫—Ü–∏—é –∫–æ–¥ –º–æ–∂–µ—Ç –ø–µ—Ä–µ–¥–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ, –æ—Ç–ª–∏—á–Ω–æ–µ –æ—Ç –Ω—É–ª—è, –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ step. –í —ç—Ç–æ–º —Å–ª—É—á–∞–µ –∫–æ–¥ –Ω–µ –±—É–¥–µ—Ç —Ä–∞–±–æ—Ç–∞—Ç—å —Ç–∞–∫, –∫–∞–∫ –ø–æ–ª–∞–≥–∞–ª–æ—Å—å.

–ü–æ–¥—Ö–æ–¥ 2. –ò—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤–ª–æ–∂–µ–Ω–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∏ –º–µ—Ö–∞–Ω–∏–∑–º –∑–∞–º—ã–∫–∞–Ω–∏–π.

```python
def draw_rect(width, height):
    def rec(step):
        if step < height:
            print('*' * width)
            rec(step + 1)

    rec(0)
```

–§—É–Ω–∫—Ü–∏—è draw_rect() –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –≤–Ω—É—Ç—Ä–∏ —Å–µ–±—è —Ä–µ–∫—É—Ä—Å–∏–≤–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é rec(), –∫–æ—Ç–æ—Ä–∞—è –∏–º–µ–µ—Ç –æ–¥–∏–Ω –ø–∞—Ä–∞–º–µ—Ç—Ä step –∏ –∏–º–µ–µ—Ç –¥–æ—Å—Ç—É–ø –∫ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º width –∏ height –≤–Ω–µ—à–Ω–µ–π —Ñ—É–Ω–∫—Ü–∏–∏. –ó–∞—Ç–µ–º —Ñ—É–Ω–∫—Ü–∏—è draw_rect() –≤—ã–∑—ã–≤–∞–µ—Ç –≤–Ω—É—Ç—Ä–µ–Ω–Ω—é—é —Ä–µ–∫—É—Ä—Å–∏–≤–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é, –ø–µ—Ä–µ–¥–∞–≤–∞—è –≤ –∫–∞—á–µ—Å—Ç–≤–µ step –∑–Ω–∞—á–µ–Ω–∏–µ 0.

–¢–µ–ø–µ—Ä—å —Ñ—É–Ω–∫—Ü–∏—è draw_rect() –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Ç–æ–ª—å–∫–æ –Ω—É–∂–Ω—ã–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã –∏ —Å–∫—Ä—ã–≤–∞–µ—Ç –≤ —Å–≤–æ–µ–π —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –∑–Ω–∞—á–µ–Ω–∏–µ step, –∫–æ—Ç–æ—Ä–æ–µ –Ω—É–∂–Ω–æ –ª–∏—à—å –¥–ª—è –æ—Å—Ç–∞–Ω–æ–≤–∫–∏ —Ä–µ–∫—É—Ä—Å–∏–∏. –¢–∞–∫–æ–π –ø–æ–¥—Ö–æ–¥ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —á–∞—Å—Ç–æ –ø—Ä–∏ –Ω–∞–ø–∏—Å–∞–Ω–∏–∏ —Ä–µ–∫—É—Ä—Å–∏–≤–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π.

–ó–∞–¥–∞—á–∞ 2. –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é print_numbers() —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Ä–µ–∫—É—Ä—Å–∏–∏, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –¥–≤–∞ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ –≤ —Å–ª–µ–¥—É—é—â–µ–º –ø–æ—Ä—è–¥–∫–µ:

start ‚Äì –Ω–∞—á–∞–ª—å–Ω–æ–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ
end ‚Äì –∫–æ–Ω–µ—á–Ω–æ–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ
–∏ –ø–µ—á–∞—Ç–∞–µ—Ç –≤—Å–µ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞ –æ—Ç start –¥–æ end –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ.

–†–µ—à–µ–Ω–∏–µ. –î–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏ –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤–ª–æ–∂–µ–Ω–Ω—É—é —Ä–µ–∫—É—Ä—Å–∏–≤–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é rec():

```python
def print_numbers(start, end):
    def rec(num):
        if num <= end:
            print(num)
            rec(num + 1)

    rec(start)


print_numbers(0, 7)
print()
print_numbers(-3, 4)
```

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 3. –ù–∞–∏–±–æ–ª–µ–µ —Ä–∞—Å–ø—Ä–æ—Å—Ç—Ä–∞–Ω–µ–Ω–Ω–æ–π –æ—à–∏–±–∫–æ–π –ø—Ä–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ —Ä–µ–∫—É—Ä—Å–∏–∏ —è–≤–ª—è–µ—Ç—Å—è –±–µ—Å–∫–æ–Ω–µ—á–Ω–∞—è —Ä–µ–∫—É—Ä—Å–∏—è, –∫–æ–≥–¥–∞ —Ü–µ–ø–æ—á–∫–∞ –≤—ã–∑–æ–≤–æ–≤ —Ñ—É–Ω–∫—Ü–∏–π –Ω–∏–∫–æ–≥–¥–∞ –Ω–µ –∑–∞–≤–µ—Ä—à–∞–µ—Ç—Å—è –∏ –ø—Ä–æ–¥–æ–ª–∂–∞–µ—Ç—Å—è, –ø–æ–∫–∞ –Ω–µ –∫–æ–Ω—á–∏—Ç—Å—è —Å–≤–æ–±–æ–¥–Ω–∞—è –ø–∞–º—è—Ç—å –≤ –∫–æ–º–ø—å—é—Ç–µ—Ä–µ. –ü—Ä–∏ —Ä–∞–∑—Ä–∞–±–æ—Ç–∫–µ —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–µ–∂–¥–µ –≤—Å–µ–≥–æ –æ—Ñ–æ—Ä–º–ª—è—Ç—å —É—Å–ª–æ–≤–∏—è –∑–∞–≤–µ—Ä—à–µ–Ω–∏—è —Ä–µ–∫—É—Ä—Å–∏–∏ –∏ –¥—É–º–∞—Ç—å, –ø–æ—á–µ–º—É —Ä–µ–∫—É—Ä—Å–∏—è –∫–æ–≥–¥–∞-–ª–∏–±–æ –∑–∞–≤–µ—Ä—à–∏—Ç —Ä–∞–±–æ—Ç—É.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 4. –í —Å–∏—Ç—É–∞—Ü–∏–∏, –∫–æ–≥–¥–∞ –Ω–µ –ø—Ä–µ–¥—É—Å–º–æ—Ç—Ä–µ–Ω —Å–ø–æ—Å–æ–± –æ—Å—Ç–∞–Ω–æ–≤–∫–∏ —Ä–µ–∫—É—Ä—Å–∏–≤–Ω—ã—Ö –≤—ã–∑–æ–≤–æ–≤, –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –ø–µ—Ä–µ–ø–æ–ª–Ω–µ–Ω–∏–µ –∞–ø–ø–∞—Ä–∞—Ç–Ω–æ–≥–æ —Å—Ç–µ–∫–∞ –∏ –≤–æ–∑–±—É–∂–¥–∞–µ—Ç—Å—è –∏—Å–∫–ª—é—á–µ–Ω–∏–µ RecursionError.

–ü–æ–¥–æ–∑—Ä–∏—Ç–µ–ª—å–Ω–æ –ø—Ä–æ—Å—Ç–æ ü§´
–í–∞–º –¥–æ—Å—Ç—É–ø–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è traffic(), —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω–Ω–∞—è —Å –ø–æ–º–æ—â—å—é —Ü–∏–∫–ª–∞ while, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ —á–∏—Å–ª–æ n –∏ –≤—ã–≤–æ–¥–∏—Ç n —Ä–∞–∑ —Å—Ç—Ä–æ–∫—É –ù–µ –ø–∞—Ä–∫–æ–≤–∞—Ç—å—Å—è.

–ü–µ—Ä–µ–ø–∏—à–∏—Ç–µ –¥–∞–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Ä–µ–∫—É—Ä—Å–∏–∏, —á—Ç–æ–±—ã –æ–Ω–∞ –≤—ã–ø–æ–ª–Ω—è–ª–∞ —Ç—É –∂–µ –∑–∞–¥–∞—á—É.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –í —Ç–µ—Å—Ç–∏—Ä—É—é—â—É—é —Å–∏—Å—Ç–µ–º—É —Å–¥–∞–π—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, —Å–æ–¥–µ—Ä–∂–∞—â—É—é —Ç–æ–ª—å–∫–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º—É—é —Ñ—É–Ω–∫—Ü–∏—é traffic(), –Ω–æ –Ω–µ –∫–æ–¥, –≤—ã–∑—ã–≤–∞—é—â–∏–π –µ–µ.

```python
def traffic(n):
    while n > 0:
        print('–ù–µ –ø–∞—Ä–∫–æ–≤–∞—Ç—å—Å—è')
        n -= 1


def traffic(n):
    if n > 0:
        traffic(n - 1)
        print('–ù–µ –ø–∞—Ä–∫–æ–≤–∞—Ç—å—Å—è')


traffic(0)
```

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ. –ö–æ–Ω—Ü–æ–º –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —è–≤–ª—è–µ—Ç—Å—è —á–∏—Å–ª–æ 0. 

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –≤–≤–µ–¥–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞ –≤ –æ–±—Ä–∞—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

```python
from sys import stdin

l = [i.strip() for i in stdin.readlines()]

def reverse(lIst: list):
    if len(lIst) == 0 or lIst[0] == '0':
        print(0)
        return

    reverse(lIst[1:])
    print(lIst[0])

reverse(l)
```

```python
def recursion():
    digit = int(input())
    if digit != 0:
        recursion()
    print(digit)

recursion()
```

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ü–∏—Ñ—Ä –≤ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–µ, —É–∫–∞–∑–∞–Ω–Ω–æ–µ –≤ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏—è—Ö, –≤—ã–≤–æ–¥–∏—Ç—å –Ω–µ –Ω—É–∂–Ω–æ.

```python
def clock(n, number):
    if n > 5:
        print(f"{str(number) * n}".center(16))
        clock(n - 4, number + 1)
        print(f"{str(number) * n}".center(16))
    else:
        print(f"{'4' * n}".center(16))

clock(16, 1)
```

–§—É–Ω–∫—Ü–∏—è print_digits() üòâ
–†–µ–∞–ª–∏–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é print_digits() —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Ä–µ–∫—É—Ä—Å–∏–∏, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –æ–¥–∏–Ω –∞—Ä–≥—É–º–µ–Ω—Ç:

number ‚Äî –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ
–§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–æ–¥–∏—Ç—å –≤—Å–µ —Ü–∏—Ñ—Ä—ã —á–∏—Å–ª–∞ number, –Ω–∞—á–∏–Ω–∞—è —Å –º–ª–∞–¥—à–∏—Ö —Ä–∞–∑—Ä—è–¥–æ–≤, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. –í —Ç–µ—Å—Ç–∏—Ä—É—é—â—É—é —Å–∏—Å—Ç–µ–º—É —Å–¥–∞–π—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, —Å–æ–¥–µ—Ä–∂–∞—â—É—é —Ç–æ–ª—å–∫–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º—É—é —Ñ—É–Ω–∫—Ü–∏—é print_digits(), –Ω–æ –Ω–µ –∫–æ–¥, –≤—ã–∑—ã–≤–∞—é—â–∏–π –µ–µ.

```python
def print_digits(number: int):
    num = str(number)
    if len(num) > 0:
        print_digits(num[1:])
        print(num[0])

print_digits(12345)
print()
print_digits(7)
```

–§—É–Ω–∫—Ü–∏—è print_digits() üòé
–†–µ–∞–ª–∏–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é print_digits() —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Ä–µ–∫—É—Ä—Å–∏–∏, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –æ–¥–∏–Ω –∞—Ä–≥—É–º–µ–Ω—Ç:

number ‚Äî –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ
–§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–æ–¥–∏—Ç—å –≤—Å–µ —Ü–∏—Ñ—Ä—ã —á–∏—Å–ª–∞ number, –Ω–∞—á–∏–Ω–∞—è —Å–æ —Å—Ç–∞—Ä—à–∏—Ö —Ä–∞–∑—Ä—è–¥–æ–≤, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. –í —Ç–µ—Å—Ç–∏—Ä—É—é—â—É—é —Å–∏—Å—Ç–µ–º—É —Å–¥–∞–π—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, —Å–æ–¥–µ—Ä–∂–∞—â—É—é —Ç–æ–ª—å–∫–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º—É—é —Ñ—É–Ω–∫—Ü–∏—é print_digits(), –Ω–æ –Ω–µ –∫–æ–¥, –≤—ã–∑—ã–≤–∞—é—â–∏–π –µ–µ.

```python
def print_digits(number: int):
    num = str(number)
    if len(num) > 0:
        print(num[0])
        print_digits(num[1:])

print_digits(12345)
```
